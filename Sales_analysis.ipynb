```python
df["Order Date"]=pd.to_datetime(df["Order Date"],dayfirst=True,errors="coerce")
# use for change the formate of date dd%mm%yy 
```

```python
df["Month"]=df["Order Date"].dt.month   # access the month
```

```python
df["Year"]=df["Order Date"].dt.year  # access the year 
```

```python
df.groupby("Month")["Sales"].sum()
```

```python
df.to_csv("C:/Sales_Project/Dataset/Cleaned_Sales_data.csv")
```
